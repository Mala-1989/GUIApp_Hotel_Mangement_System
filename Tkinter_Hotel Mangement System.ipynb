```python
'''
In this project I have created GUI App used for hotel management system. Its consist of


Frame1:--Take the order from the customer include-
Meals
Drinks
Booking room
Order delivery
Currency Convertor2
 
Frame2Contain 4 Buttons
Total->To get total of Order
Screen->Blank the screen of total but entries remain intact
Reset->Reset all the entries
Exit->To close the app.
 
Frame3
Contains the calculator
Display Scroll  Function

'''
```

```python
from tkinter import *
import time
from tkinter import ttk
import datetime

msg1 = "Please make your order..."
text1 = ""
n = 0
#---------------------------Calculating funtions------------------#
def btn(numbers):
    global operator
    operator = operator + str(numbers)
    txt_input.set(operator)

def Clear():
    global operator
    operator = ''
    txt_input.set('')
    Display.insert(0,'Start Calculating..')

def Equal():
    global operator
    sumup = float(eval(operator))
    txt_input.set(sumup)
    operator = ''
#---------------------------Total funtions------------------#

def TotalResult():
    #-------------------Cost of Meal--------------#
    varme1 = Mealdicator.get()
    varme2 = Meal1.get()
    if varme1 == 'Rice':
        varme3 = (varme2*1.8)
        Cost.set(varme3)
    elif varme1 == 'Cheese':
        varme3 = (varme2 * 3.7)
        Cost.set(varme3)
    elif varme1 == 'Fried Rice Chicken':
        varme3 = (varme2 * 4.5)
        Cost.set(varme3)
    elif varme1 == 'Sharwame':
        varme3 = (varme2 * 2.5)
        Cost.set(varme3)
    elif varme1 == 'Burger':
        varme3 = (varme2 * 1.0)
        Cost.set(varme3)
    else:
        varme3 = (varme2 * 0.0)
        Cost.set(varme3)

    #-------------------Cost of Drinks--------------#

    verdi1 = Drinkdicator.get()
    vardi2 = Drink1.get()
    if verdi1 == 'Bottled Water':
        vardi3 = (vardi2 * 0.5)
        Drinks.set(vardi3)
    elif verdi1 == 'Cococola':
        vardi3 = (vardi2 * 1.0)
        Drinks.set(vardi3)
    elif verdi1 == 'Pepsi':
        vardi3 = (vardi2 * 1.0)
        Drinks.set(vardi3)
    elif verdi1 == 'Strawberry Shakes':
        vardi3 = (vardi2 * 2.0)
        Drinks.set(vardi3)
    elif verdi1 == 'Red Wine':
        vardi3 = (vardi2 * 3.5)
        Drinks.set(vardi3)
    else:
        vardi3 = (vardi2 * 0.0)
        Drinks.set(vardi3)

    # -------------------Delievery Cost-------------#

    num1 = float(Cost.get())
    num2 = float(Drinks.get())
    Delivery = (num1 + num2) * 0.2

    #--------------------Booking Room----------------#

    room = v.get()
    null = 0.0

    rvip = 10.0 # Cost of VIP Room#
    rvip1 = Delivery/(10*0.5)#-VIP Room with Delivery cost-#

    rnormal = 5.0 #-Normal Room Cost-#
    rnormal1 = Delivery/(5*2.5)#-Normal Room with Delivery cost-#

    if room == 1:
        if chkb1.get() == 1:
            Service_Charge.set(rvip)
            RoomCost.set(10.0)
            DevCost.set(Delivery)
        else:
            Service_Charge.set(null)
            DevCost.set(null)
            RoomCost.set(10.0)

    elif room == 2:
        if chkb1.get() == 2:
            Service_Charge.set(rnormal)
            RoomCost.set(5.0)
            DevCost.set(Delivery)
        else:
            Service_Charge.set(null)
            DevCost.set(null)
            RoomCost.set(10.0)

    elif room == 3:
        if chkb1.get() == 1:
            Service_Charge.set(null)
            RoomCost.set(null)
            DevCost.set(null)
        else:
            Service_Charge.set(null)
            DevCost.set(null)
            RoomCost.set(null)


 #-----------------------Total Result------------------#

    num3 = float(DevCost.get())
    num4 = float(RoomCost.get())
    num5 = float(Service_Charge.get())

    MyTotal = num1+num2+num3+num4+num5
    Total.set(MyTotal)
    FinalTotal = "Total = $",MyTotal

    num6 = Total.get()
    Display.delete(0,END)
    Display.insert(0,FinalTotal)

    if num6 == '0.0':
        Display.delete(0,END)
        Display.insert(0, 'Please make an order..')

    # -------------------Convert Funtion--------------#
def Convert():
    var2 = indicator.get()
    var3 = var1.get()
    if var2 == "China":
        Display.delete(0,END)
        var4 = ('Yaun',(var3*6.98))
        Display.insert(0,var4)
    elif var2 == "France":
        Display.delete(0,END)
        var4 = ('Euro',(var3*0.81))
        Display.insert(0,var4)
    elif var2 == "India":
        Display.delete(0,END)
        var4 = ('Rupees',(var3/68.8))
        Display.insert(0,var4)
    elif var2 == "UK":
        Display.delete(0,END)
        var4 = ('Pound',(var3*0.82))
        Display.insert(0,var4)
    elif var2 == "USA":
        Display.delete(0,END)
        var4 = ('USD',(var3*1.00))
        Display.insert(0,var4)
    else:
        Display.delete(0,END)
        Display.insert(0, 'Error:select a counry')

    #----------------------RESET Button--------------#
def Hotel():
    Display.delete(0,END)
    Display.insert(0,'Hotel Management Sys.')

def Powered():
    Display.delete(0,END)
    Display.insert(0,'Powered by Python..')

def Reset():
    Display.delete(0,END)
    Display.insert(0,'System Resetting..')
    Display.after(2000,Hotel)
    Display.after(4000,Powered)
    Display.after(6000,Rest)

def Rest():
    Clear()
    Display.delete(0,END)
    Display.insert(0,'Hello! Welcome')
    Mealdicator.set(value='Delicios Meal')
    Drinkdicator.set(value='Fresh Drink')
    indicator.set(value='Choose a country')
    txtQtyofMeal.delete(0,END)
    txtQtyofMeal.insert(0,0)
    txtQtyofDrink.delete(0,END)
    txtQtyofDrink.insert(0,0)
    txtTotal.delete(0,END)
    txtTotal.insert(0,0)
    RoomCost.set(0.0)
    Total.set(0.0)
    Service_Charge.set(0.0)
    Cost.set(0.0)
    Drinks.set(0.0)
    chkb1.set(0)
    v.set(3)
    DevCost.set(0.0)

    #-------------------------Clear Button----------------------#
def ClearScreen():
    Display.delete(0,END)
    RoomCost.set('')
    Total.set('')
    Service_Charge.set('')
    Cost.set('')
    Drinks.set('')
    DevCost.set('')

    #-------------------Exit--------------#
def Stop():
    root.destroy()

def Exit():
    Display.delete(0,END)
    Display.insert(0,'Thanks for Patronage...')
    Display.after(3000,Stop)

    #-------------------Time-------------------#

def Tick():
    d = datetime.datetime.now()
    Today = '(:%B %d, %Y)'.format(d)

    mytime = time.strftime('%I:%M:%S%p')
    lblInfo.config(text=(mytime+''+Today))
    lblInfo.after(200,tick)

    #-----------------Scroll Funtion-----------------#

def display():
    global text1, n, msg1
    for t in range(len(msg1)):
        for k in range(t):
            text1 += ' '
        for g in range(len(msg1)-t):
            text1 += msg1[g]
        text1 = text1.strip()
        f2.update()
        f2.after(200)
        Scroll_txt.set('')
        Scroll_txt.set(text1)
        text1 = ''
    Scroll_txt.set('')
    txtscroll.after(200,display)



root = Tk()
root.geometry('1600x800+0+0')
root.title('Hotel Management Syetem')


#---------------------------Windows Partition------------------#
Tops = Frame(root, width=1800,height=100, bg='blue',relief=SUNKEN)
Tops.pack(side=TOP)

f1 = Frame(root,width=500, height=700)
f1.pack(side=LEFT)

f2 = Frame(root, width=100, height=700)
f2.pack(side=RIGHT)

f3 = Frame(root, width=35, height=700)
f3.pack(side=LEFT)

f4 = Frame(root, width=70, height=700)
f4.pack(side=LEFT)


#---------------------------Main Screen------------------#

txt_input = StringVar(value = 'Welcome..!!')
Display = Entry(Tops, font=('arial',80,'bold'),fg='white',bd=15,bg='blue',justify='right',textvariable=txt_input)
Display.grid(columnspan=4)

#--------------------------Date and Time-----------------#

localtime = time.asctime(time.localtime(time.time()))

lblInfo = Label(f2,font=('arial',20,'bold'),text=localtime,fg='dark blue',bd=4,anchor=W)
lblInfo.grid(row=1,column=0,columnspan=4)


#---------------------------Build the Calculator------------------#

#-----------------ROW-----------------#
operator=''

btn7 = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='7',command=lambda:btn(7)).grid(row=2,column=0)
btn8 = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='8',command=lambda:btn(8)).grid(row=2,column=1)
btn9 = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='9',command=lambda:btn(9)).grid(row=2,column=2)
btnC = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='C',bg='green',command=Clear).grid(row=2,column=3)

#-----------------ROW2----------------#
btn4 = Button(f2,padx=15,pady=2,bd=8,fg='black',font=('arial',25,'bold'),text ='4',command=lambda:btn(4)).grid(row=3,column=0)
btn5 = Button(f2,padx=15,pady=2,bd=8,fg='black',font=('arial',25,'bold'),text ='5',command=lambda:btn(5)).grid(row=3,column=1)
btn6 = Button(f2,padx=15,pady=2,bd=8,fg='black',font=('arial',25,'bold'),text ='6',command=lambda:btn(6)).grid(row=3,column=2)
btnplus = Button(f2,padx=18,pady=2,bd=8,fg='black',font=('arial',25,'bold'),text ='+',bg='orange',command=lambda:btn('+')).grid(row=3,column=3)


#-----------------ROW3----------------#

btn1 = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='1',command=lambda:btn(1)).grid(row=4,column=0)
btn2 = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='2',command=lambda:btn(2)).grid(row=4,column=1)
btn3 = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='3',command=lambda:btn(3)).grid(row=4,column=2)
btnminus = Button(f2,padx=23,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='-',bg='orange',command=lambda:btn('-')).grid(row=4,column=3)

#-----------------ROW4----------------#

btn0 = Button(f2,padx=15,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='0',command=lambda:btn(0)).grid(row=5,column=0)
btndot = Button(f2,padx=21,pady=5,bd=8,fg='black',font=('arial',25,'bold'),text ='.',bg='orange',command=lambda:btn('.')).grid(row=5,column=1)
btndivision = Button(f2,padx=20,pady=4,bd=8,fg='black',font=('arial',25,'bold'),text ='/',bg='orange',command=lambda:btn('/')).grid(row=5,column=2)
btnmultiply = Button(f2,padx=19,pady=4,bd=8,fg='black',font=('arial',25,'bold'),text ='x',bg='orange',command=lambda:btn('*')).grid(row=5,column=3)

#-----------------ROW5----------------#

btnequal = Button(f2,padx=64,pady=2,bd=8,fg='black',font=('arial',25,'bold'),text ='=',bg='green',command=Equal).grid(row=6,column=0,columnspan=2)
btnopenbracket = Button(f2,padx=19,pady=4,bd=8,fg='black',font=('arial',25,'bold'),text ='(',bg='orange',command=lambda:btn('(')).grid(row=6,column=2)
btnclosebracket = Button(f2,padx=23,pady=4,bd=8,fg='black',font=('arial',25,'bold'),text =')',bg='orange',command=lambda:btn(')')).grid(row=6,column=3)


#---------------------------Choose Meal------------------#

Meal1 = IntVar()
Mealdicator = StringVar(value='Delicious Meal')

lblMeal = Label(f1,font=('arial',13,'bold'),text='Choose Meal',bd=5,anchor=W)
lblMeal.grid(row=0,column=0)
txtMeal = ttk.Combobox(f1,font=('arial',13,'bold'),textvariable = Mealdicator)
txtMeal['values']=('Rice', 'Cheese','Fried Rice Chicken','Sharwame','Burger')
txtMeal.grid(row=0,column=1)

lblQtofMeal = Label(f1,font=('arial',13,'bold'),text = 'Quantity of Meal',bd=5,anchor=W)
lblQtofMeal.grid(row=1,column=0)
txtQtyofMeal = Entry(f1, font=('arial',13,'bold'),textvariable=Meal1,bd=5,insertwidth=3,bg='white',justify='right')
txtQtyofMeal.grid(row=1,column=1)

#---------------------------Choose Drinks------------------#

Drink1 = IntVar()
Drinkdicator = StringVar(value='Fresh Drinks')

lblDrink = Label(f1,font=('arial',13,'bold'),text='Choose Drinks',bd=5,anchor=W)
lblDrink.grid(row=2,column=0)
txtDrink = ttk.Combobox(f1,font=('arial',13,'bold'),textvariable = Drinkdicator)
txtDrink['values']=('Bottled Water', 'Cococola','Pepsi','Strawberry Shakes','Red Wine')
txtDrink.grid(row=2,column=1)


lblQtofDrink = Label(f1,font=('arial',13,'bold'),text = 'Quantity of Drink',bd=5,anchor=W)
lblQtofDrink.grid(row=3,column=0)
txtQtyofDrink = Entry(f1, font=('arial',13,'bold'),textvariable=Drink1,bd=5,insertwidth=2,bg='white',justify='right')
txtQtyofDrink.grid(row=3,column=1)

#---------------------------order delivery------------------#
chkb1 = IntVar()

lblHomeDev = Label(f1, font=('arial',13,'bold'),text='Order Delivery',bd=5,anchor=W)
lblHomeDev.grid(row=4,column=0)
check1 = Checkbutton(f1,text='Yes', variable=chkb1,font=('arial',13,'bold'))
check1.grid(row=4,column=1)

#---------------------------Book a room------------------#
v = IntVar()

v.set(3)
lblRoom = Label(f1, font=('arial',13,'bold'),text='Book a Room',bd=5,anchor=W)
lblRoom.grid(row=5,column=0)
MyRadios = Radiobutton(f1, font=('arial',13,'bold'),text='VIP',variable=v,value=1)
MyRadios.grid(row=5,column=1,sticky=W)
MyRadios = Radiobutton(f1, font=('arial',13,'bold'),text='Normal',variable=v,value=2)
MyRadios.grid(row=5,column=1)
MyRadios = Radiobutton(f1, font=('arial',13,'bold'),text='No',variable=v,value=3)
MyRadios.grid(row=5,column=1,sticky=E)

#---------------------------Cost Display Screen-----------------#

Cost = StringVar()

lblMeal1 = Label(f1,font=('arial',13,'bold'),text='Cost of meal($)',bd=5,anchor=W)
lblMeal1.grid(row=0,column=2)
txtMeal1 = Entry(f1,font=('arial',13,'bold'),textvariable = Cost,fg='white',bd=5,insertwidth=3,bg='blue',justify='right')
txtMeal1.grid(row=0,column=3)


Drinks = StringVar()
lblDrink1 = Label(f1,font=('arial',13,'bold'),text='Cost of drink($)',bd=5,anchor=W)
lblDrink1.grid(row=1,column=2)
txtDrink1 = Entry(f1,font=('arial',13,'bold'),textvariable = Drinks,fg='white',bd=5,insertwidth=3,bg='blue',justify='right')
txtDrink1.grid(row=1,column=3)


DevCost = StringVar()

lblDev1 = Label(f1,font=('arial',13,'bold'),text='Delivery Cost($)',bd=5,anchor=W)
lblDev1.grid(row=2,column=2)
txtDev1 = Entry(f1,font=('arial',13,'bold'),textvariable = DevCost,fg='white',bd=5,insertwidth=3,bg='blue',justify='right')
txtDev1.grid(row=2,column=3)


RoomCost = StringVar()

lblRoom = Label(f1,font=('arial',13,'bold'),text='Cost of Room($)',bd=5,anchor=W)
lblRoom.grid(row=3,column=2)
txtRoom = Entry(f1,font=('arial',13,'bold'),textvariable = RoomCost,fg='white',bd=5,insertwidth=3,bg='blue',justify='right')
txtRoom.grid(row=3,column=3)


Service_Charge = StringVar()

lblService = Label(f1,font=('arial',13,'bold'),text='Service Fee($)',bd=5,anchor=W)
lblService.grid(row=4,column=2)
txtService = Entry(f1,font=('arial',13,'bold'),textvariable = Service_Charge,fg='white',bd=5,insertwidth=3,bg='blue',justify='right')
txtService.grid(row=4,column=3)

Total = StringVar()

lblTotal = Label(f1,font=('arial',13,'bold'),text='Total Cost($)',bd=5,anchor=W)
lblTotal.grid(row=5,column=2)
txtTotal = Entry(f1,font=('arial',13,'bold'),textvariable = Total,fg='white',bd=5,insertwidth=3,bg='blue',justify='right')
txtTotal.grid(row=5,column=3)

#---------------------------Currency Converter----------------#

var1 = IntVar()

indicator = StringVar(value='Choose a Country')

lblCunCon = Label(f1,font=('arial',13,'bold italic'),text='-------------------------------------Currency Converter-----------------------------------',
                  bd=5,anchor=W)
lblCunCon.grid(row=6,column=0,columnspan=4)

lblCountry = Label(f1,font=('arial',13,'bold'),text='Natonality',bd=5,anchor=W)
lblCountry.grid(row=7,column=0)
txtCountry = ttk.Combobox(f1,font=('arial',13,'bold'),textvariable=indicator)
txtCountry['values'] = ('China','India','USA','UK','France')
txtCountry.grid(row=7,column=1)

lblAmount = Label(f1,font=('arial',13,'bold'),text='Amount($)',bd=5,anchor=W)
lblAmount.grid(row=7,column=2)
textAmount = Entry(f1,font=('arial',13,'bold'),textvariable=var1,bd=5,insertwidth=3,bg='blue',justify='right')
textAmount.grid(row=7,column=3)

#--------------------------Control Buttons----------------#

btnConvert = Button(f1,padx=5,pady=4,bd=8,fg='white',font=('arial',16,'bold'),width=10,text='Convert',bg='orange',command=Convert)
btnConvert.grid(row=8,column=2)

btnTotal = Button(f3,padx=5,pady=8,bd=8,fg='white',font=('arial',16,'bold'),width=10,text='Total',bg='orange',command=TotalResult)
btnTotal.grid(row=0,column=2)

btnScreen = Button(f3,padx=5,pady=8,bd=8,fg='white',font=('arial',16,'bold'),width=10,text='Screen',bg='blue',command=ClearScreen)
btnScreen.grid(row=1,column=2)

btnReset = Button(f3,padx=5,pady=8,bd=8,fg='white',font=('arial',16,'bold'),width=10,text='Reset',bg='green',command=Reset)
btnReset.grid(row=2,column=2)

btnExit = Button(f3,padx=5,pady=8,bd=8,fg='white',font=('arial',16,'bold'),width=10,text='Exit',bg='red',command=Exit)
btnExit.grid(row=3,column=2)

#--------------------------Scrollable Text----------------#

Scroll_txt = StringVar()
txtscroll = Entry(f2, textvariable = Scroll_txt,font=('arial',16,'bold'),fg='white',bd=5,bg='blue',width=32)
txtscroll.grid(row=0,column=0,columnspan=4)
display()

root.mainloop()
```
